```python
from unsloth import FastLanguageModel
import torch

# Р—Р°РіСЂСѓР¶Р°РµРј РЅР°С€Сѓ РїСЂРµРґРѕР±СѓС‡РµРЅРЅСѓСЋ РјРѕРґРµР»СЊ-СѓС‡РёС‚РµР»СЏ
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Llama-3.2-3B-Instruct-bnb-4bit",
    max_seq_length = 2048,
    load_in_4bit = True,
)
```

```python
# Р”РѕР±Р°РІР»СЏРµРј Р°РґР°РїС‚РµСЂ РґР»СЏ СЌС„С„РµРєС‚РёРІРЅРѕРіРѕ РѕР±СѓС‡РµРЅРёСЏ
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
)
print("вњ… LoRA Р°РґР°РїС‚РµСЂ РЅР°СЃС‚СЂРѕРµРЅ!")
```

```python
from datasets import load_dataset

print("Р—Р°РіСЂСѓР·РєР° РґР°С‚Р°СЃРµС‚Р° GrandMaster-PRO-MAX...")
dataset = load_dataset("Vikhrmodels/GrandMaster-PRO-MAX", split="train")
print(f"рџ“љ Р’СЃРµРіРѕ РїСЂРёРјРµСЂРѕРІ РІ РґР°С‚Р°СЃРµС‚Рµ: {len(dataset)}")

# Р¤РёР»СЊС‚СЂСѓРµРј С‚РѕР»СЊРєРѕ СЂСѓСЃСЃРєРёРµ РїСЂРёРјРµСЂС‹
filtered_dataset = dataset.filter(lambda x: x['prompt_lang'] == 'ru' and x['answer_lang'] == 'ru')
print(f"вњ… Р СѓСЃСЃРєРёС… РїСЂРёРјРµСЂРѕРІ: {len(filtered_dataset)}")

# Р‘РµСЂС‘Рј РЅРµР±РѕР»СЊС€СѓСЋ С‡Р°СЃС‚СЊ РґР»СЏ РѕР±СѓС‡РµРЅРёСЏ (РЅР°С‡РЅРёС‚Рµ СЃ 500-1000)
small_dataset = filtered_dataset.select(range(1000))
print(f"рџ“– Р”Р»СЏ РѕР±СѓС‡РµРЅРёСЏ РІР·СЏС‚Рѕ: {len(small_dataset)} РїСЂРёРјРµСЂРѕРІ")
```

```python
def format_instruct(example):
    messages = example['conversation']
    return {"text": tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False
    )}

formatted_dataset = small_dataset.map(format_instruct)
print("вњ… Р”Р°РЅРЅС‹Рµ РѕС‚С„РѕСЂРјР°С‚РёСЂРѕРІР°РЅС‹!")

# РџРѕРєР°Р¶РµРј РїСЂРёРјРµСЂ, РєР°Рє Р±СѓРґРµС‚ РІС‹РіР»СЏРґРµС‚СЊ РІС…РѕРґ РґР»СЏ РјРѕРґРµР»Рё
print("\nрџ“ќ РџСЂРёРјРµСЂ С„РѕСЂРјР°С‚РёСЂРѕРІР°РЅРёСЏ:")
print(formatted_dataset[0]['text'][:400] + "...")
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = formatted_dataset,
    dataset_text_field = "text",
    max_seq_length = 2048,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        num_train_epochs = 1,  # Р”Р»СЏ 1000 РїСЂРёРјРµСЂРѕРІ РґРѕСЃС‚Р°С‚РѕС‡РЅРѕ 1 СЌРїРѕС…Рё
        learning_rate = 2e-4,
        fp16 = True,
        output_dir = "./output",
        logging_steps = 5,
        save_steps = 50,
        report_to = "none",  # РћС‚РєР»СЋС‡Р°РµРј РѕС‚РїСЂР°РІРєСѓ РѕС‚С‡С‘С‚РѕРІ
    ),
)

print("рџљЂ РќР°С‡РёРЅР°РµРј РѕР±СѓС‡РµРЅРёРµ...")
trainer.train()
print("вњ… РћР±СѓС‡РµРЅРёРµ Р·Р°РІРµСЂС€РµРЅРѕ!")
```

```python
# РўРµСЃС‚РёСЂСѓРµРј РјРѕРґРµР»СЊ РЅР° РїСЂРѕСЃС‚РѕРј РїСЂРёРјРµСЂРµ
prompt = "Р§С‚Рѕ РѕР·РЅР°С‡Р°РµС‚ СЃР»РѕРІРѕ 'run' РІ РєРѕРЅС‚РµРєСЃС‚Рµ 'He runs a company'?"
messages = [{"role": "user", "content": prompt}]
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")

outputs = model.generate(inputs, max_new_tokens=200, temperature=0.7)
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("рџ¤– Р’РѕРїСЂРѕСЃ:", prompt)
print("рџ“– РћС‚РІРµС‚ РјРѕРґРµР»Рё:", response[len(prompt)+10:])  # РћС‚РґРµР»СЏРµРј РѕС‚РІРµС‚ РѕС‚ РїСЂРѕРјРїС‚Р°
```
